# Sales Performance Over Time

```python
#  2024 total sales
df_totalsales2024 = df_sales.groupby(df_sales["sale_date"].dt.to_period("Y"))["total_price"].sum()
df_totalsales2024
```

```python
#target of 2024 
sales = df_sales['total_price'].sum()
target = df_sales_targets['sales_target'].sum()

achievement = (sales / target) * 100

print(f"Total Sales :{sales}")
print(f"Target :{target}")
print(f"Achievement :{achievement}")
```

```python
!pip install pygal
!pip install cairosvg
import pygal
from IPython.display import SVG
```

```python
from pygal.style import Style

custom_style = Style(
    background='transparent',
    plot_background='transparent',
    foreground='#000',
    foreground_strong='#000',
    foreground_subtle='#555',
    opacity='.6',
    opacity_hover='.9',
    transition='400ms ease-in',
    colors=('#5cb85c',) if achievement >= 100 else ('#f0ad4e',)

)
# solid gauge
gauge = pygal.SolidGauge(
    inner_radius=0.70,
    style=custom_style
)


percent_formatter = lambda x: '{:.1f}%'.format(x)
gauge.value_formatter = percent_formatter


gauge.add('Target Achievement', [{'value': achievement, 'max_value': 100}])
```

```python
# year sales for branchs
df_yearsales = df_sales.groupby(df_sales["branch_id"])["total_price"].sum()
df_yearsales
```

```python
#Target Achievement
df_targets = df_sales_targets.groupby('branch_id', as_index=False)['sales_target'].sum()


df_merged = pd.merge(df_yearsales, df_targets, on='branch_id')


df_merged['achievement'] = (df_merged['total_price'] / df_merged['sales_target']) * 100


def get_style(achievement):
    color = '#5cb85c' if achievement >= 100 else '#f0ad4e'
    return Style(
        background='transparent',
        plot_background='transparent',
        foreground='#000',
        foreground_strong='#000',
        foreground_subtle='#555',
        opacity='.6',
        opacity_hover='.9',
        transition='400ms ease-in',
        colors=(color,)
    )


for _, row in df_merged.iterrows():
    branch = row['branch_id']
    achievement = row['achievement']

    gauge = pygal.SolidGauge(
        half_pie=True,
        inner_radius=0.70,
        style=get_style(achievement),
        title=f"Branch {branch}"
    )
    gauge.value_formatter = lambda x: '{:.1f}%'.format(x)
    gauge.add('Target Achievement', [{'value': achievement, 'max_value': 100}])

    display(SVG(gauge.render()))
```

```python
# sales per month
df_monthlysales = df_sales.groupby(df_sales["sale_date"].dt.to_period("M"))["total_price"].sum()
df_monthlysales
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Plot sales trend
plt.figure(figsize=(12,6))
df_monthlysales.plot(kind="line", marker="o", color="b")
plt.title("Monthly Sales Trend")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.grid()
plt.show()
```

```python
#Branch Sales
df_Bmonthlysales = df_sales.groupby([pd.Grouper(key='sale_date', freq='M'), 'branch_id'], as_index=False)['total_price'].sum()


df_pivot = df_Bmonthlysales.pivot(index='sale_date', columns='branch_id', values='total_price')

df_pivot
```

```python
plt.figure(figsize=(12,6))
sns.set(style="whitegrid")


df_pivot.plot(marker='o', figsize=(12,6))
plt.title("Monthly Sales Trend per Branch", fontsize=16)
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.legend(title='Branch')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Best-Selling Shoes

```python
# sales by Shoes category
df_merged = pd.merge(df_sales, df_shoes, on='shoe_id')
top_shoes = df_merged.groupby('category')['total_price'].sum().sort_values(ascending=False)

top_shoes
```

```python
# Plot Top shoes
plt.figure(figsize=(8,5))
sns.barplot(x=top_shoes.values, y=top_shoes.index, palette="Blues_r")
plt.title("Top-Selling by Shoes category")
plt.xlabel("Total Sales")
plt.show()
```

# Customer Preferences

```python
#  sales by Shoes color
df_merged = pd.merge(df_sales, df_shoes, on='shoe_id')
shoes_color = df_merged.groupby('color')['total_price'].sum().sort_values(ascending=False)

shoes_color
```

```python
# Plot shoes Color
plt.figure(figsize=(8,5))
sns.barplot(x=shoes_color.values, y=shoes_color.index, palette="Blues_r")
plt.title("Top-Selling by Shoes color")
plt.xlabel("Total Sales")
plt.show()
```

```python
#  sales by Shoes size
df_merged = pd.merge(df_sales, df_shoes, on='shoe_id')
shoes_size = df_merged.groupby('size')['total_price'].sum().sort_values(ascending=False)

shoes_size
```

```python
# Plot shoes size
plt.figure(figsize=(8,5))
sns.barplot(x=shoes_size.index, y=shoes_size.values)
plt.title("Top-Selling by Shoes size")
plt.xlabel("Total Sales")
plt.show()
```
